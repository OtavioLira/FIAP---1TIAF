```python
#Importação das bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_breast_cancer
```

```python
# Carrega o dataset de cancer de mama de winsconsin
breast = load_breast_cancer()

# Convertendo para o dataset do pandas(dataset, columns)
df = pd.DataFrame(breast.data, columns = breast.feature_names)

df["label"] = breast.target

# Exibindo 10 exemplos aleatorios para leitura da base de dados
df.head(10)
```

```python
# Separar dados e label
from sklearn.model_selection import train_test_split

X = df.drop(columns=["label"])
y = df["label"]

# Separar treino e teste
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=101
)
```

```python
# Padronizar os dados
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Reduzindo a dimensionalidade com PCA
from sklearn.decomposition import PCA

n_components = 5  # Escolha os componentes principais como preferir
pca = PCA(n_components=n_components)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
# Criação do modelo preditivo de Regressão Logistica
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(solver="liblinear")
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
```

```python
# Avaliação do modelo de Regressão Logistica
from sklearn import metrics

print("     ACC: ", metrics.accuracy_score(y_test, y_pred))
print("Precisão: ", metrics.precision_score(y_test, y_pred))
print("  Recall: ", metrics.recall_score(y_test, y_pred))
```

```python
# Criação do modelo preditivo SVM
from sklearn import svm

model = svm.SVC(kernel="rbf", C=1.0)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
# Criação do modelo preditivo SVM
print("     ACC: ", metrics.accuracy_score(y_test, y_pred))
print("Precisão: ", metrics.precision_score(y_test, y_pred))
print("  Recall: ", metrics.recall_score(y_test, y_pred))
```

```python
# Criação do modelo preditivo Arvore de Decisão
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)
y_pred = tree.predict(X_test)
```

```python
# Criação do modelo preditivo Arvore de Decisão
print("     ACC: ", metrics.accuracy_score(y_test, y_pred))
print("Precisão: ", metrics.precision_score(y_test, y_pred))
print("  Recall: ", metrics.recall_score(y_test, y_pred))
```
