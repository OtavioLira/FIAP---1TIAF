```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans, DBSCAN
```

```python
url = "https://raw.githubusercontent.com/reisanar/datasets/master/moons.csv"
df = pd.read_csv(url)
df
```

```python
# Plotar os pontos
plt.scatter(x = df["X"], y = df["Y"])
plt.show()
```

# Criando e comparando clusters

```python
# Define os clusters
km = KMeans(n_clusters = 4, random_state = 42)
db = DBSCAN(eps = 0.4)

# Verifica a distribuiÃ§Ã£o dos pontos
c_km = km.fit_predict(df)
c_db = db.fit_predict(df)
```

```python
c_km
```

```python
c_db
```

```python
# Visualizando os clusters pelos mÃ©todos distintos
fig = plt.figure( figsize = (12,8) )

plt.subplot(2, 2, 1)
plt.title("Dados originais")
plt.scatter(x = df["X"], y = df["Y"])

plt.subplot(2,2,3) #linhas, colunas, quadrado( inferior esquerdo )
plt.title("K-Means")
plt.scatter(x = df["X"], y = df["Y"], c = c_km)


plt.subplot(2,2,4) #linhas, colunas, quadrado(4 - inferior direito)
plt.title("DBSCAN")
plt.scatter(x = df["X"], y = df["Y"], c = c_db)

plt.show()
```

```python
df = pd.read_csv("mall.csv")
df.sample(10)
```

```python
df.info()
```

```python
sns.barplot(df)
sns.pairplot(df)
```

```python
# Filtrar as variaveis de interesse
df_f = df[ ["Annual Income (k$)", "Spending Score (1-100)"] ]
```

# KMeans
Encontrando o melhor valor de K (mÃ©todo do cotovelo)

```python
erros = []
for k in range(2, 11):
  kmeans = KMeans(n_clusters = k, random_state = 42)
  kmeans.fit(df_f)
  erros.append( kmeans.inertia_ )

plt.plot(list( range(2, 11) ), erros, "-o" )
plt.xlabel = ("Valores de K")
plt.ylabel("SSE")
plt.title("Escolha o valor de k pelo mÃ©todo do cotovelo")
plt.show()
```

### ClusterizaÃ§Ã£o final por KMeans

```python
# inicializa o algoritmo
km = KMeans(n_clusters = 5, random_state = 42)

# Cria os clusters
km.fit(df_f)

# Obtem os centroids
centroids = km.cluster_centers_

# Obtem a clusterizaÃ§Ã£o de cada dado
km_labels = km.predict(df_f)
```

```python
# -> clusters
plt.scatter(df_f["Annual Income (k$)"],       # Eixo X
            df_f["Spending Score (1-100)"],   # Eixo y
            alpha = 0.5,                      # Nivel de transparencia de 50%
            c = km_labels)                    # Esqueam de cores (1 cor para cluster)

# -> centroids
plt.scatter(centroids[:,0],         # Eixo X
            centroids[:,1],         # Eixo y
            c =  "black",           # Todos os centroids serÃ£o pretos
            marker = "X",           # Tipo de marcador como um X #https://matplotlib.org/stable/api/markers_api.html#module-matplotlib.markers
            s = 200)                # Tamanho do marcador
```

# ClusterizaÃ§Ã£o por DBSCAN

```python
# InicializaÃ§Ã£o do algoritmo
db = DBSCAN(eps = 10, min_samples = 8)

# Criar os cluster
db.fit(df_f)

# ObtÃ©m a clusterizaÃ§Ã£o de cada dado
db_labels = db.labels_
db_labels
```

```python
# -> clusters
plt.scatter(df_f["Annual Income (k$)"],       # Eixo X
            df_f["Spending Score (1-100)"],   # Eixo y
            alpha = 0.5,                      # Nivel de transparencia de 50%
            c = db_labels)                    # Esqueam de cores (1 cor para cluster)
```

```python
# -> clusters, DBSCAN, Sem os outliers
mascara = db_labels >= 0
plt.scatter(df_f["Annual Income (k$)"][mascara],       # Eixo X
            df_f["Spending Score (1-100)"][mascara],   # Eixo y
            alpha = 0.5,                               # Nivel de transparencia de 50%
            c = db_labels[mascara])                    # Esqueam de cores (1 cor para cluster)

plt.ylabel("Score de gastos (0-100)")
```

```python
df_f["Cluster_kmeans"] = km_labels
df_f["Cluster_dbscan"] = db_labels
```
