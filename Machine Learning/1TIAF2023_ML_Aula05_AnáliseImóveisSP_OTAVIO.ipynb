### Exercício

- Filtre imóveis em SP nas seguintes condições:
    - Preço até 2000,00
    - Que possuam elevador
    - Com metragem acima de 85 metros quadrados
    - e com ao menos 03 quartos

- Quais são os 5 maiores apartamentos? DICA: veja a função _nlargest()_

- Quais são os 10 apartamentos mais caros? E em caso de "empate", quais possuem o maior tamanho?

- Em relação ao local (District), deixe o texto em CAPSLOCK

```python
mascaraPreco = imoveisSP["Price"] > 2000
mascaraElevador = imoveisSP["Elevator"] == True
mascaraMetragem = imoveisSP["Size"] > 85
mascaraQuartos = imoveisSP["Rooms"] >= 3
```

```python
# Filtro na ordem das linhas
imoveisSP = imoveisSP[mascaraPreco & mascaraElevador & mascaraMetragem & mascaraQuartos]
imoveisSP
```

```python
# Filtro, ordenado pelo preço
imoveisSP["Price"].sort_values(ascending=False)
```

```python
# Top 5 aptos maiores
imoveisSP["Size"].nlargest(5) #.sort_values().tail(5)
```

```python
# Quais são os 10 apartamentos mais caros? E em caso de "empate", quais possuem o maior tamanho?
#mascaraValor = imoveisSP["Price"].sort_values(ascending=False).nlargest(10)
#mascaraTamanho = imoveisSP["Size"].sort_values
imoveisSP.nlargest(10, columns=["Price","Size"])
```
